### (a)
**Assume that $d=1$ and $n=1$. The sample is the $n$ of size 1 and boils down to just $(x, y)$. The function $F$ writes simply as**
$$F(\beta) = (y-x\beta)^2 + \lambda\beta^2.$$
**Compute and write down the gradient $\nabla F$ of $F$.**



$\space\space\space\space\nabla F$

$= 2(y-x\beta)(0-x) + 2\lambda \beta\space\space\space\space$(Apply Chain rule)

$= -2x(y-x\beta)+2\lambda \beta$

### (b)
**Assume now that $d > 1$ and $n > 1$. Using the previous result and the linearity of differentiation, compute and write down the gradient $\nabla F(\beta)$ of $F$.**

We have
$$F(\beta) = \frac{1}{n} \sum_{i=1}^n (y_i-x_i^T\beta)^2 + \lambda\mid\mid\beta\mid\mid_2^2$$

We can write
$$F(\beta) = g(\beta) + h(\beta)$$

$$ where \space g(\beta) = \frac{1}{n} \sum_{i=1}^n (y_i-x_i^T\beta)^2, h(\beta) = \lambda\mid\mid\beta\mid\mid_2^2$$

So the gradient $\nabla F(\beta)$ of $F$ can be written as
$$\nabla F(\beta) = \frac{\delta g}{\delta\beta} + \frac{\delta h}{\delta\beta}$$

We can first derive $\frac{\delta g}{\delta\beta}$. Notice that we can rewrite
$$g(\beta) = \frac{1}{n} \sum_{i=1}^n (y_i-x_i^T\beta)^2$$

$$= \frac{1}{n} \mid\mid Y-X^T\beta \mid\mid_2^2$$

$$= \frac{1}{n}\{(Y-X^T\beta)^T(Y-X^T\beta)\}$$

$$= \frac{1}{n}\{Y^T Y - Y^T X^T \beta - (X^T \beta)^T Y + (X^T \beta)^T X^T \beta))\}$$

$$= \frac{1}{n}\{Y^T Y - Y^T X^T \beta - \beta ^T XY + \beta ^T XX^T \beta\}$$

So we have
$$\frac{\delta g}{\delta\beta} = \frac{1}{n} \{ \frac{\delta}{\delta\beta}(Y^T Y) - \frac{\delta}{\delta\beta}(Y^T X^T \beta) - \frac{\delta}{\delta\beta}(\beta ^T XY) + \frac{\delta}{\delta\beta}(\beta ^T XX^T \beta) \}$$

Using the Hammer Identity, for any $A$, $$\frac{\delta}{\delta x}(A^T x) = \frac{\delta}{\delta x}(x^T A) = A,$$

we have

$$\frac{\delta}{\delta\beta}(Y^T Y) = 0$$

$$\frac{\delta}{\delta\beta}(Y^T X^T \beta) = \frac{\delta}{\delta\beta}((XY)^T \beta) = \frac{\delta}{\delta\beta}(\beta ^T (XY)) = XY$$

$$\frac{\delta}{\delta\beta}(\beta ^T XY) = \frac{\delta}{\delta\beta}(\beta ^T (XY)) = XY$$

$$\frac{\delta}{\delta\beta}(\beta ^T XX^T \beta) = \frac{\delta}{\delta\beta}\beta ^T (XX^T) \beta = (XX^T + 
(XX^T)^T)\beta = (XX^T + XX^T)\beta = 2XX^T \beta$$

Therefore,
$$\frac{\delta g}{\delta\beta}
= \frac{1}{n}\{0 - XY - XY + 2XX^T \beta\}
= \frac{1}{n}\{-2XY + 2XX^T \beta\}
= -\frac{2}{n}X(Y-X^T \beta)$$

Now let's derive $\frac{\delta h}{\delta\beta}$ using the Hammer Identity, for any $A$, $$\frac{\delta(X^T AX)}{\delta X} = (A+A^T)X$$

Then we get
$$\frac{\delta h}{\delta\beta}
= \frac{\delta(\lambda\mid\mid \beta \mid\mid_2^2)}{\delta\beta}
= \lambda\frac{\delta(\mid\mid \beta \mid\mid_2^2)}{\delta\beta}
= \lambda\frac{\delta(\beta^T \beta)}{\delta\beta}
= \lambda(I + I^T)\beta
= 2\lambda\beta$$

Thus,
$$\nabla F(\beta)
= \frac{\delta g}{\delta\beta} + \frac{\delta h}{\delta\beta}
= -\frac{2}{n}X(Y-X^T \beta) + 2\lambda\beta$$

### (c)
**Consider the `Hitters` dataset, which you should load and divide into training and test sets using the code below.**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

# Load the data
hitters = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Hitters.csv', sep=',', header=0)
hitters = hitters.dropna()

# Create our X matrix with the predictors and y vector with the response
X = hitters.drop('Salary', axis=1)
X = pd.get_dummies(X, drop_first=True)
y = hitters.Salary

# Divide the data into training and test sets. By default, 25% goes into the test set.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

**Standardize the data. Note that you can convert a data frame into an array by using `np.array()`.**

```python
import numpy as np

# Standardize the data
X_train = preprocessing.scale(np.array(X_train)).T
y_train = preprocessing.scale(np.array(y_train)).T
X_test = preprocessing.scale(np.array(X_test)).T
y_test = preprocessing.scale(np.array(y_test)).T
```

### (d)
**Write a function `computegrad` that computes and returns $\nabla F(\beta)$ for any $\beta$. Avoid using for loops by vectorizing the computation.**

```python
def computegrad(X,Y,beta,lamb):
    """
    Compute and returns gradient
    """
    n = X.shape[1]
    return (-(2/n)*X.dot(Y-X.T.dot(beta))+2*lamb*beta)
```

### (e)
**Write a function `graddescent` that implements the gradient descent algorithm described in Algorithm 1. The function `graddescent` calls the function `computegrad` as a sub-routine. The function takes as input the initial point, the constant step-size value, and the maximum number of iterations. The stopping criterion is the maximum number of iterations.**

```python
def graddescent(X, Y, init, eta, max_iter, lamb):
    """
    Gradient Descent Algorithm
    input: step-size ùúÇ
    initialization: ùõΩ0 = 0
    repeat for t = 0,1,2,...
           ùõΩt+1 = ùõΩt - ùúÇ * ‚àáF(ùõΩ)
    until the stoppeing criterion
    """
    beta = init
    grad = computegrad(X, Y, beta, lamb)
    vals = [beta]
    t = 0
    # The stopping criterion is max_iter
    while t < max_iter:
        beta = beta - eta * grad
        vals.append(beta) # save beta
        grad = computegrad(X, Y, beta, lamb)
        t = t + 1
    return np.array(vals)
```

### (f)
**Set the constant step-size to $\eta$ = 0.05 and the maximum number of iterations to 1000. Run `graddescent` on the training set of the `Hitters` dataset for $\lambda$ = 0.05. Plot the curve of the objective value $F(\beta _t)$ versus the iteration counter $t$. (Again, avoid using for loops when computing the objective values). What do you observe?**

```python
def obj(X,Y,beta,lamb):
    """
    Compute objective value
    """
    n = X.shape[1]
    return 1/n*np.linalg.norm(Y-X.T.dot(beta))**2+lamb*np.linalg.norm(beta)**2
```

```python
import matplotlib.pyplot as plt

def obj_plot(X,Y,x_vals,lamb):
    """
    Plot objection value over iterations
    """
    n = x_vals.shape[0] # number of iterations
    fs = np.zeros(n)
    for i in range(n):
        fs[i] = obj(X,Y,x_vals[i],lamb)
    plt.figure()
    plt.plot(fs)
    plt.xlabel('Iteration')
    plt.ylabel('Objective Value')
    plt.title('Objective Value over Iterations', fontsize=16)
    plt.show()
    return
```

```python
# Set the constant step-size to eta = 0.05
# Set the maximum number of iterations to max_iter = 1000
# Run graddescent on the training set of Hitters for lambda = 0.05
vals = graddescent(X_train, y_train, init=np.zeros(19,), eta=0.05, max_iter=1000, lamb=0.05)
```

```python
# Plot
obj_plot(X_train, y_train, x_vals=vals, lamb=0.05)
```

From the plot of the objective value versus the iteration counter, we can see the objective value $F(\beta _t)$ is decreasing as the iteration counter $t$ increases. This satisfy our optimization goal to minimize the objective value $F(\beta _t)$.

### (g)
**Denote $\beta _T$ the final iterate of your gradient descent algorithm. Compare $\beta _T$ to the $\beta ^*$ found by `sklearn.linear model.Ridge`. Compare the objective value for $\beta _T$ to the one for $\beta ^*$. What do you observe?**

```python
def obj_ridge(X,Y,beta,lamb):
    """
    Compute objective value
    """
    return np.linalg.norm(X.T.dot(beta)-Y)**2+lamb*np.linalg.norm(beta)**2
```

```python
from sklearn.linear_model import Ridge

# Find beta_T by gradient sedcent algorithm
vals = graddescent(X_train, y_train, init=np.zeros(19,), eta=0.05, max_iter=1000, lamb=0.05)
beta_T = vals[vals.shape[0]-1]
beta_T_obj = obj(X_train, y_train, beta_T, lamb=0.05)

# Find beta_star by Ridge model
model = Ridge(alpha=0.05)
ridge = model.fit(X_train.T, y_train.T)
beta_star = model.coef_
beta_star_obj = obj_ridge(X_train, y_train, beta_star, lamb=0.05)

print('beta_T =')
print(beta_T)
print('\nbeta^* =')
print(beta_star)
print('\nThe objective value for beta_T = %.8f' % beta_T_obj)
print('\nThe objective value for beta^* = %.8f' % beta_star_obj)
```

The $\beta_T$ is different from $\beta^*$. The objective value for $\beta_T$ is much smaller than the objective value for $\beta^*$.

Notice that for the gradient descent algorithm, the optimization problem minimize:

$$\frac{1}{n}\mid\mid Y-X^T \beta \mid\mid_2^2 + \lambda\mid\mid\beta\mid\mid_2^2$$

in which $\beta_T$ achieves the minimized objective value.

For Ridge regression, the ridge coefficients minimize a penalized residual sum of squares (From ocumentation: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html)

$$\mid\mid Y-X^T \beta \mid\mid_2^2 + \lambda\mid\mid\beta\mid\mid_2^2$$

in which $\beta^*$ achieves the minimized objective value.

### (h)
**Run your gradient algorithm for many values of $\eta$ on a logarithmic scale. Find the final iterate, across all runs for all the values of $\eta$, that achieves the smallest value of the objective. Compare $\beta _T$ to the $\beta ^*$ found by `sklearn.linear_model.Ridge`. Compare the objective value for $\beta _T$ to the $\beta ^*$. What conclusion to you draw?**

```python
# Run for different eta on a lafarithmic scale
# Find beta_T with smallest objective by gradient sedcent algorithm
min_obj = 10000
count = 1
while count < 4:
    eta = np.log(count) # choose different eta on a logarithmic scale
    vals = graddescent(X_train, y_train, init=np.zeros(19,), eta=eta, max_iter=1000, lamb=0.05)
    beta = vals[vals.shape[0]-1]
    beta_obj = obj(X_train, y_train, beta, lamb=0.05)
    if beta_obj < min_obj:
        min_obj = beta_obj # update minimum objective value
        beta_T = beta # update beta_T
    count += 0.01
beta_T_obj = min_obj

# Find beta_star by Ridge model
model = Ridge(alpha=0.05)
ridge = model.fit(X_train.T, y_train.T)
beta_star = model.coef_
beta_star_obj = obj_ridge(X_train, y_train, beta_star, lamb=0.05)

print('beta_T =')
print(beta_T)
print('\nbeta^* =')
print(beta_star)
print('\nThe objective value for beta_T = %.8f' % beta_T_obj)
print('\nThe objective value for beta^* = %.8f' % beta_star_obj)
```

For the same reason as above, the $\beta_T$ is different from $\beta^*$. The objective value for $\beta_T$ is much smaller than the objective value for $\beta^*$.

We can draw a conclusion here: For all $\eta$ values run on a logarithmic scale, we can find a final iterate $\beta_T$ that achieves the smallest value of the objective. In this problem, the final iterate for the gradient descent algorithm gives a smaller objective value than the final iterate for Ridge regression model because these two algorithms solve different optimization problem.

### (b)
**Use the `OLS` function from the `statsmodels` package to perform a simple linear regression with `mpg` as the response and `weight` as the predictor. Be sure to include an intercept. Use the `summary()` attribute to print the results. Comment on the output.**

**Hint: See this URL for help with the `statsmodels` functions: http://www.statsmodels.org/dev/regression.html#examples**

```python
import statsmodels.api as sm
X = sm.add_constant(auto['weight'])
model = sm.OLS(auto['mpg'], X)
results = model.fit()
print(results.summary())
```


**(i) Is there a relationship between the predictor and the response?**
    
The p-value of the F-statistic is 6.02e-102, which is pretty closed to zero, so we reject the null hypothesis. There is a relationship between weight and mpg.

The R-squared 0.693 indicates that 69.3% of the variance in mpg is due to weight.

**(iii) Is the relationship between the predictor and the response positive or negative?**

Notice that the coefficient -0.0076 is negative, so the relationship between weight and mpg is negative.

```python
import statsmodels as sm
plot = sm.graphics.regressionplots.plot_fit(results, exog_idx=1)
```

**This problem involves simple linear regression without an intercept.**

### (a)
**Recall that the coefficient estimate $\beta$ for the linear regression of $Y$ onto $X$ without an intercept is given by (3.38). Under what circumstance is the coefficient estimate for the regression of $X$ onto $Y$ the same as the coefficient estimate for the regression of $Y$ onto $X$?**

The coefficient estimate $\beta$ for the linear regression of $Y$ onto $X$ without an intercept is given by

$$\hat \beta = \frac {\sum_{i=1}^n x_i y_i} {\sum_{i^\prime=1}^n x_{i^\prime}^2} $$

The coefficient estimate $\beta$ for the linear regression of $X$ onto $Y$ without an intercept is given by

$$\hat \beta = \frac {\sum_{i=1}^n x_i y_i} {\sum_{i^\prime=1}^n y_{i^\prime}^2} $$

So the coefficient estimate for the regression of $X$ onto $Y$ the same as the coefficient estimate for the regression of $Y$ onto $X$ when $\sum_{i^\prime=1}^n x_{i^\prime}^2 = \sum_{i^\prime=1}^n y_{i^\prime}^2$.

### (b)
**Generate an example in Python with $n = 50$ observations in which the coefficient estimate for the regression of $X$ onto $Y$ is different from the coefficient estimate for the regression of $Y$ onto $X$.**

```python
import numpy as np
import pandas as pd

np.random.seed(0)
# Set x to be [1,2,3,...,49,50]
x = np.array(range(1,51))
X = pd.DataFrame(x, columns=['X'])

# Set y to be x plus normally randomly picked 50 difference
y = x + np.random.normal(loc=0,size=50)
Y = pd.DataFrame(y, columns=['Y'])
```

```python
sum(x ** 2)
```

```python
sum(y ** 2)
```

In this case, $\sum_{i^\prime=1}^n x_{i^\prime}^2 \neq \sum_{i^\prime=1}^n y_{i^\prime}^2$. So the coefficient estimate for the regression of $X$ onto $Y$ is supposed to be different from the coefficient estimate for the regression of $Y$ onto $X$.

```python
import statsmodels.api as sm
model1 = sm.OLS(X,Y)
result1 = model1.fit()
print(result1.summary())
```

```python
model2 = sm.OLS(Y,X)
result2 = model2.fit()
print(result2.summary())
```

The coefficient estimate for the regression of $X$ onto $Y$ is 1.0013.

The coefficient estimate for the regression of $Y$ onto $X$ is 0.9972.

They are different.
